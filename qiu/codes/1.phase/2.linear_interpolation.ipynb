```python
import pandas as pd
import numpy as np
import hvplot.pandas
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
%matplotlib inline
```

```python
def get_delta(time):
    delta = 10+time.minute%30
    return time - pd.Timedelta(minutes=delta)
```

```python
data_path = '../../../data/'
to_pred = [276264, 276265]
```

```python
tti = pd.read_csv(data_path + 'train_TTI.csv')
tti['time'] = pd.to_datetime(tti['time'])
tti['tod'] = tti['time'].dt.hour*100+tti['time'].dt.minute
print(len(tti))
```

```python
tti=tti[(tti['tod']>=730)&(tti['tod']<=2130)]
tti['weekday'] = tti['time'].dt.weekday
tti['x_ts'] = tti.apply(lambda row: get_delta(row['time']),axis=1)
tti.drop(columns=['speed'],inplace=True)
tti.tail()
```

```python
tti1=pd.merge(tti,tti[['id_road','time','TTI']],left_on=['id_road','x_ts'],right_on=['id_road','time'])
tti1.drop(columns=['x_ts'],inplace=True)
tti1.tail(10)
```

```python
linear={}
for i1, road in enumerate(tti1['id_road'].unique()):
    for weekday in range(7):
        ttirw = tti1[(tti1['id_road']==road)&(tti1['weekday']==weekday)]
        for i2, tod in enumerate(ttirw['tod'].unique()):
            ttirwt = ttirw[ttirw['tod']==tod]
            x=LinearRegression()
            x.fit(ttirwt[['TTI_x']],ttirwt[['TTI_y']])
            linear[(road,weekday,tod)]=x
```
